```python
"""
knowledge_base.py — Knowledge Base E-Commerce Platform (Single File)
"""

import json

# ============================================================
# DATA
# ============================================================

KB = {
    "platform": "E-Commerce Platform",
    "version": "1.0",
    "entities": [
        {
            "id": "ecommerce",
            "name": "E-Commerce Platform",
            "type": "platform",
            "type_label": "Platform Utama",
            "description": "Marketplace digital untuk produk elektronik dan lainnya, beroperasi di skala enterprise dalam domain retail online.",
            "properties": {
                "fungsi": "Marketplace digital produk elektronik & lainnya",
                "skala": "Platform level enterprise",
                "domain": "Retail online"
            },
            "relations": [
                {"label": "memuat",            "direction": "out", "target_id": "laptop"},
                {"label": "memiliki kategori", "direction": "out", "target_id": "elektronik"},
                {"label": "digunakan oleh",    "direction": "out", "target_id": "pengguna"},
            ]
        },
        {
            "id": "pengguna",
            "name": "Pengguna",
            "type": "aktor",
            "type_label": "Aktor Utama",
            "description": "Pembeli dan pencari produk yang masuk melalui E-Commerce Platform.",
            "properties": {
                "peran": "Pembeli & pencari produk",
                "aktivitas_utama": "Membeli, mencari produk berdasarkan kebutuhan",
                "entry_point": "E-Commerce Platform"
            },
            "relations": [
                {"label": "membeli",           "direction": "out", "target_id": "laptop"},
                {"label": "membeli",           "direction": "out", "target_id": "smartphone"},
                {"label": "membeli",           "direction": "out", "target_id": "smart_tv"},
                {"label": "mencari berdasar",  "direction": "out", "target_id": "kerja"},
                {"label": "menggunakan",       "direction": "out", "target_id": "ecommerce"},
            ]
        },
        {
            "id": "elektronik",
            "name": "Elektronik",
            "type": "kategori",
            "type_label": "Kategori Produk",
            "description": "Kategori induk (top-level) yang mencakup semua perangkat elektronik di platform.",
            "properties": {
                "level": "Kategori induk (top-level)",
                "cakupan": "Semua perangkat elektronik",
                "sub_kategori": ["Smartphone", "Smart TV", "Laptop"]
            },
            "relations": [
                {"label": "mencakup",      "direction": "out", "target_id": "laptop"},
                {"label": "mencakup",      "direction": "out", "target_id": "smartphone"},
                {"label": "mencakup",      "direction": "out", "target_id": "smart_tv"},
                {"label": "kategori dari", "direction": "in",  "target_id": "ecommerce"},
            ]
        },
        {
            "id": "laptop",
            "name": "Laptop",
            "type": "produk",
            "type_label": "Produk",
            "description": "Perangkat komputasi portabel untuk segmen consumer dan profesional, berada di rentang harga mid–high.",
            "properties": {
                "tipe": "Perangkat komputasi portabel",
                "segmen": "Consumer & profesional",
                "harga": "Mid–high range"
            },
            "relations": [
                {"label": "kategori",          "direction": "out", "target_id": "elektronik"},
                {"label": "sesuai kebutuhan",  "direction": "out", "target_id": "kerja"},
                {"label": "dibeli oleh",       "direction": "in",  "target_id": "pengguna"},
            ]
        },
        {
            "id": "smartphone",
            "name": "Smartphone",
            "type": "produk",
            "type_label": "Produk",
            "description": "Perangkat mobile komunikasi berbasis Android atau iOS.",
            "properties": {
                "tipe": "Perangkat mobile komunikasi",
                "os": "Android / iOS",
                "induk_kategori": "Elektronik"
            },
            "relations": [
                {"label": "bagian dari", "direction": "in", "target_id": "elektronik"},
                {"label": "dibeli oleh", "direction": "in", "target_id": "pengguna"},
            ]
        },
        {
            "id": "smart_tv",
            "name": "Smart TV",
            "type": "produk",
            "type_label": "Produk",
            "description": "Televisi berbasis internet dan sistem operasi dengan fitur streaming, browser, dan aplikasi.",
            "properties": {
                "tipe": "Televisi berbasis internet & OS",
                "fitur": ["Streaming", "Browser", "Apps"],
                "induk_kategori": "Elektronik"
            },
            "relations": [
                {"label": "bagian dari", "direction": "in", "target_id": "elektronik"},
                {"label": "dibeli oleh", "direction": "in", "target_id": "pengguna"},
            ]
        },
        {
            "id": "kerja",
            "name": "Kerja",
            "type": "kebutuhan",
            "type_label": "Kebutuhan / Konteks",
            "description": "Konteks penggunaan profesional dengan intent produktivitas dan mobilitas.",
            "properties": {
                "tipe_konteks": "Konteks penggunaan profesional",
                "sub_konteks": ["Kantor", "Remote"],
                "intent_user": "Produktivitas & mobilitas"
            },
            "relations": [
                {"label": "memiliki konteks", "direction": "out", "target_id": "kantor"},
                {"label": "memiliki konteks", "direction": "out", "target_id": "remote"},
                {"label": "dicari oleh",      "direction": "in",  "target_id": "pengguna"},
            ]
        },
        {
            "id": "kantor",
            "name": "Kantor",
            "type": "konteks",
            "type_label": "Sub-Konteks",
            "description": "Lingkungan kerja stasioner yang membutuhkan laptop/PC, monitor, dan aksesori pendukung.",
            "properties": {
                "tipe": "Lingkungan kerja stasioner",
                "kebutuhan_perangkat": ["Laptop/PC", "Monitor", "Aksesori"],
                "induk": "Kerja"
            },
            "relations": [
                {"label": "sub-konteks dari", "direction": "in",  "target_id": "kerja"},
                {"label": "relevan untuk",    "direction": "out", "target_id": "laptop"},
            ]
        },
        {
            "id": "remote",
            "name": "Remote",
            "type": "konteks",
            "type_label": "Sub-Konteks",
            "description": "Konteks bekerja dari luar kantor (WFH/WFA) yang membutuhkan laptop portabel, headset, dan webcam.",
            "properties": {
                "tipe": "Bekerja dari luar kantor (WFH/WFA)",
                "kebutuhan_perangkat": ["Laptop portabel", "Headset", "Webcam"],
                "induk": "Kerja"
            },
            "relations": [
                {"label": "sub-konteks dari", "direction": "in",  "target_id": "kerja"},
                {"label": "relevan untuk",    "direction": "out", "target_id": "laptop"},
            ]
        },
    ]
}

# ============================================================
# LOADER
# ============================================================

def get_entity_by_id(entity_id: str) -> dict | None:
    for e in KB["entities"]:
        if e["id"] == entity_id:
            return e
    return None

def get_entities_by_type(entity_type: str) -> list:
    return [e for e in KB["entities"] if e["type"] == entity_type]

def get_related_entities(entity_id: str) -> list:
    entity = get_entity_by_id(entity_id)
    if not entity:
        return []
    result = []
    for rel in entity.get("relations", []):
        related = get_entity_by_id(rel["target_id"])
        if related:
            result.append({
                "relation_label": rel["label"],
                "direction": rel["direction"],
                "entity": related
            })
    return result

# ============================================================
# QUERY
# ============================================================

def query_by_type(entity_type: str) -> list:
    entities = get_entities_by_type(entity_type)
    return [{"id": e["id"], "name": e["name"], "type_label": e["type_label"]} for e in entities]

def query_entity_detail(entity_id: str) -> dict | None:
    entity = get_entity_by_id(entity_id)
    if not entity:
        return None
    related = get_related_entities(entity_id)
    return {
        "id": entity["id"],
        "name": entity["name"],
        "type": entity["type"],
        "type_label": entity["type_label"],
        "description": entity["description"],
        "properties": entity.get("properties", {}),
        "relations": [
            {
                "label": r["relation_label"],
                "direction": r["direction"],
                "target_id": r["entity"]["id"],
                "target_name": r["entity"]["name"],
            }
            for r in related
        ],
    }

def query_relations_of(entity_id: str, direction: str = "all") -> list:
    entity = get_entity_by_id(entity_id)
    if not entity:
        return []
    rels = entity.get("relations", [])
    if direction != "all":
        rels = [r for r in rels if r["direction"] == direction]
    result = []
    for r in rels:
        target = get_entity_by_id(r["target_id"])
        result.append({
            "label": r["label"],
            "direction": r["direction"],
            "target_id": r["target_id"],
            "target_name": target["name"] if target else r["target_id"],
        })
    return result

def query_path(from_id: str, to_id: str) -> list:
    if not get_entity_by_id(from_id) or not get_entity_by_id(to_id):
        return []
    queue = [[(from_id, None, None)]]
    visited = {from_id}
    while queue:
        path = queue.pop(0)
        current_id = path[-1][0]
        if current_id == to_id:
            return [{"entity_id": s[0], "via_label": s[1], "direction": s[2]} for s in path]
        if len(path) >= 6:
            continue
        entity = get_entity_by_id(current_id)
        if not entity:
            continue
        for rel in entity.get("relations", []):
            nxt = rel["target_id"]
            if nxt not in visited:
                visited.add(nxt)
                queue.append(path + [(nxt, rel["label"], rel["direction"])])
    return []

def query_search(keyword: str) -> list:
    keyword_lower = keyword.lower()
    results = []
    for entity in KB["entities"]:
        hit = (
            keyword_lower in entity["name"].lower()
            or keyword_lower in entity.get("description", "").lower()
        )
        if not hit:
            for v in entity.get("properties", {}).values():
                val_str = " ".join(v) if isinstance(v, list) else str(v)
                if keyword_lower in val_str.lower():
                    hit = True
                    break
        if hit:
            results.append({
                "id": entity["id"],
                "name": entity["name"],
                "type_label": entity["type_label"],
                "description": entity["description"],
            })
    return results

# ============================================================
# FORMATTER
# ============================================================

def format_entity(entity: dict) -> str:
    if not entity:
        return "Entitas tidak ditemukan."
    lines = [
        f"{'=' * 50}",
        f"  {entity['name'].upper()}  |  {entity['type_label']}",
        f"{'=' * 50}",
        f"\nDeskripsi:\n  {entity['description']}\n",
        "Properti:",
    ]
    for key, value in entity.get("properties", {}).items():
        display_val = ", ".join(value) if isinstance(value, list) else value
        lines.append(f"  - {key.replace('_', ' ').title()}: {display_val}")
    lines += ["\nRelasi Semantik:"]
    for rel in entity.get("relations", []):
        arrow = "→" if rel["direction"] == "out" else "←"
        lines.append(f"  {arrow} [{rel['label']}] {rel['target_name']}")
    lines.append("")
    return "\n".join(lines)

def format_entity_list(entities: list) -> str:
    if not entities:
        return "Tidak ada entitas ditemukan."
    lines = [f"Ditemukan {len(entities)} entitas:\n"]
    for i, e in enumerate(entities, 1):
        lines.append(f"  {i}. [{e['type_label']}] {e['name']}")
    return "\n".join(lines)

def format_path(path: list) -> str:
    if not path:
        return "Tidak ada jalur yang ditemukan."
    id_to_name = {e["id"]: e["name"] for e in KB["entities"]}
    parts = []
    for i, step in enumerate(path):
        name = id_to_name.get(step["entity_id"], step["entity_id"])
        if i == 0:
            parts.append(name)
        else:
            parts.append(f"--[{step['via_label']}]--> {name}")
    return "Jalur ditemukan:\n  " + " ".join(parts)

def format_search_results(keyword: str, results: list) -> str:
    if not results:
        return f"Tidak ada entitas yang cocok dengan keyword '{keyword}'."
    lines = [f"Hasil pencarian '{keyword}' — {len(results)} entitas:\n"]
    for e in results:
        lines.append(f"  [{e['type_label']}] {e['name']}")
        lines.append(f"    {e['description']}\n")
    return "\n".join(lines)

# ============================================================
# MAIN / DEMO
# ============================================================

if __name__ == "__main__":
    print("\n" + "#" * 60)
    print("  KNOWLEDGE BASE — E-COMMERCE PLATFORM")
    print("#" * 60)

    print("\n--- Semua Entitas ---")
    all_e = [{"id": e["id"], "name": e["name"], "type_label": e["type_label"]} for e in KB["entities"]]
    print(format_entity_list(all_e))

    print("\n--- Detail: Laptop ---")
    print(format_entity(query_entity_detail("laptop")))

    print("\n--- Produk yang tersedia ---")
    print(format_entity_list(query_by_type("produk")))

    print("\n--- Relasi keluar dari Elektronik ---")
    rels = query_relations_of("elektronik", direction="out")
    for r in rels:
        print(f"  → [{r['label']}] {r['target_name']}")

    print("\n--- Jalur: Pengguna → Kantor ---")
    print(format_path(query_path("pengguna", "kantor")))

    print("\n--- Pencarian keyword: 'portabel' ---")
    print(format_search_results("portabel", query_search("portabel")))
```

```python
"""
ecommerce_semantic.py
Sistem E-Commerce berbasis Knowledge Base Semantik
Jalankan: python ecommerce_semantic.py
"""

# ============================================================
# KNOWLEDGE BASE DATA
# ============================================================

KB = {
    "entities": [
        {
            "id": "ecommerce",
            "name": "E-Commerce Platform",
            "type": "platform",
            "description": "Marketplace digital untuk produk elektronik.",
            "properties": {"domain": "Retail online", "skala": "Enterprise"},
        },
        {
            "id": "pengguna",
            "name": "Pengguna",
            "type": "aktor",
            "description": "Pembeli dan pencari produk.",
            "properties": {"peran": "Pembeli", "aktivitas": "Membeli & mencari produk"},
        },
        {
            "id": "elektronik",
            "name": "Elektronik",
            "type": "kategori",
            "description": "Kategori induk semua perangkat elektronik.",
            "properties": {"level": "Top-level", "sub_kategori": ["Laptop", "Smartphone", "Smart TV"]},
        },
        {
            "id": "laptop",
            "name": "Laptop",
            "type": "produk",
            "description": "Perangkat komputasi portabel untuk kerja dan produktivitas.",
            "properties": {"harga": 12000000, "stok": 50, "kategori": "Elektronik", "cocok_untuk": ["kerja", "kantor", "remote"]},
        },
        {
            "id": "smartphone",
            "name": "Smartphone",
            "type": "produk",
            "description": "Perangkat komunikasi mobile berbasis Android/iOS.",
            "properties": {"harga": 5000000, "stok": 120, "kategori": "Elektronik", "cocok_untuk": ["komunikasi", "hiburan"]},
        },
        {
            "id": "smart_tv",
            "name": "Smart TV",
            "type": "produk",
            "description": "Televisi pintar dengan fitur streaming dan internet.",
            "properties": {"harga": 8000000, "stok": 30, "kategori": "Elektronik", "cocok_untuk": ["hiburan", "rumah"]},
        },
        {
            "id": "kerja",
            "name": "Kerja",
            "type": "kebutuhan",
            "description": "Konteks kebutuhan profesional dan produktivitas.",
            "properties": {"intent": "Produktivitas", "sub_konteks": ["kantor", "remote"]},
        },
        {
            "id": "kantor",
            "name": "Kantor",
            "type": "konteks",
            "description": "Lingkungan kerja stasioner di kantor.",
            "properties": {"tipe": "Stasioner", "induk": "kerja"},
        },
        {
            "id": "remote",
            "name": "Remote / WFH",
            "type": "konteks",
            "description": "Bekerja dari rumah atau luar kantor.",
            "properties": {"tipe": "Fleksibel", "induk": "kerja"},
        },
    ],
    "relations": [
        {"from": "laptop",     "label": "kategori",         "to": "elektronik"},
        {"from": "smartphone", "label": "kategori",         "to": "elektronik"},
        {"from": "smart_tv",   "label": "kategori",         "to": "elektronik"},
        {"from": "laptop",     "label": "sesuai kebutuhan", "to": "kerja"},
        {"from": "laptop",     "label": "cocok untuk",      "to": "kantor"},
        {"from": "laptop",     "label": "cocok untuk",      "to": "remote"},
        {"from": "pengguna",   "label": "membeli",          "to": "laptop"},
        {"from": "pengguna",   "label": "membeli",          "to": "smartphone"},
        {"from": "pengguna",   "label": "membeli",          "to": "smart_tv"},
        {"from": "pengguna",   "label": "mencari berdasar", "to": "kerja"},
        {"from": "kerja",      "label": "memiliki konteks", "to": "kantor"},
        {"from": "kerja",      "label": "memiliki konteks", "to": "remote"},
    ],
    "keranjang": [],
    "riwayat_pembelian": [],
}


# ============================================================
# HELPER
# ============================================================

def get_entity(entity_id: str) -> dict | None:
    for e in KB["entities"]:
        if e["id"] == entity_id:
            return e
    return None

def get_products() -> list:
    return [e for e in KB["entities"] if e["type"] == "produk"]

def get_relations_from(entity_id: str) -> list:
    return [r for r in KB["relations"] if r["from"] == entity_id]

def format_harga(harga: int) -> str:
    return f"Rp {harga:,.0f}".replace(",", ".")

def separator(char="─", width=55):
    print(char * width)

def header(title: str):
    separator("═")
    print(f"  {title}")
    separator("═")


# ============================================================
# FITUR: LIHAT SEMUA PRODUK
# ============================================================

def lihat_produk():
    header("DAFTAR PRODUK")
    produk_list = get_products()
    for i, p in enumerate(produk_list, 1):
        stok = p["properties"]["stok"]
        harga = p["properties"]["harga"]
        status = "✓ Tersedia" if stok > 0 else "✗ Habis"
        print(f"  {i}. {p['name']}")
        print(f"     Harga : {format_harga(harga)}")
        print(f"     Stok  : {stok} unit  |  {status}")
        print(f"     Info  : {p['description']}")
        separator("·")


# ============================================================
# FITUR: REKOMENDASI SEMANTIK
# ============================================================

def rekomendasi():
    header("REKOMENDASI PRODUK")
    print("  Masukkan kebutuhanmu dan sistem akan merekomendasikan")
    print("  produk yang paling sesuai secara semantik.\n")

    kebutuhan_input = input("  Kebutuhanmu (contoh: kerja, hiburan, komunikasi): ").strip().lower()

    cocok = []
    for produk in get_products():
        cocok_untuk = produk["properties"].get("cocok_untuk", [])
        # Cek kecocokan semantik: langsung atau via relasi konteks
        if any(kebutuhan_input in c for c in cocok_untuk):
            cocok.append(produk)
        elif kebutuhan_input in produk["description"].lower():
            cocok.append(produk)
        elif kebutuhan_input in produk["name"].lower():
            cocok.append(produk)

    # Cek via relasi kebutuhan (misal input "kantor" → cari produk yang relasi ke kantor)
    if not cocok:
        for r in KB["relations"]:
            target = get_entity(r["to"])
            if target and kebutuhan_input in target["name"].lower():
                sumber = get_entity(r["from"])
                if sumber and sumber["type"] == "produk" and sumber not in cocok:
                    cocok.append(sumber)

    separator()
    if cocok:
        print(f"  Produk yang cocok untuk '{kebutuhan_input}':\n")
        for p in cocok:
            print(f"  ► {p['name']} — {format_harga(p['properties']['harga'])}")
            print(f"    {p['description']}")
    else:
        print(f"  Tidak ada produk yang cocok untuk '{kebutuhan_input}'.")
        print("  Coba kata kunci: kerja, kantor, remote, hiburan, komunikasi")
    separator()


# ============================================================
# FITUR: CARI PRODUK
# ============================================================

def cari_produk():
    header("CARI PRODUK")
    keyword = input("  Masukkan nama produk: ").strip().lower()

    hasil = [
        e for e in KB["entities"]
        if keyword in e["name"].lower() or keyword in e["description"].lower()
    ]

    separator()
    if hasil:
        for e in hasil:
            tipe = e["type"].upper()
            print(f"  [{tipe}] {e['name']}")
            print(f"    {e['description']}")
            if e["type"] == "produk":
                print(f"    Harga : {format_harga(e['properties']['harga'])}")
                print(f"    Stok  : {e['properties']['stok']} unit")
            separator("·")
    else:
        print(f"  Produk '{keyword}' tidak ditemukan.")
    separator()


# ============================================================
# FITUR: TAMBAH KE KERANJANG
# ============================================================

def tambah_keranjang():
    header("TAMBAH KE KERANJANG")
    produk_list = get_products()

    for i, p in enumerate(produk_list, 1):
        print(f"  {i}. {p['name']} — {format_harga(p['properties']['harga'])} (stok: {p['properties']['stok']})")

    separator()
    try:
        pilihan = int(input("  Pilih nomor produk (0 = batal): "))
        if pilihan == 0:
            return
        if not 1 <= pilihan <= len(produk_list):
            print("  Pilihan tidak valid.")
            return

        produk = produk_list[pilihan - 1]
        if produk["properties"]["stok"] == 0:
            print(f"  Stok {produk['name']} habis!")
            return

        jumlah = int(input(f"  Jumlah {produk['name']}: "))
        if jumlah <= 0:
            print("  Jumlah tidak valid.")
            return
        if jumlah > produk["properties"]["stok"]:
            print(f"  Stok tidak cukup. Stok tersedia: {produk['properties']['stok']}")
            return

        # Cek apakah sudah ada di keranjang
        for item in KB["keranjang"]:
            if item["id"] == produk["id"]:
                item["jumlah"] += jumlah
                print(f"\n  ✓ {produk['name']} x{jumlah} ditambahkan ke keranjang.")
                return

        KB["keranjang"].append({"id": produk["id"], "name": produk["name"], "harga": produk["properties"]["harga"], "jumlah": jumlah})
        print(f"\n  ✓ {produk['name']} x{jumlah} ditambahkan ke keranjang.")

    except ValueError:
        print("  Input tidak valid.")


# ============================================================
# FITUR: LIHAT KERANJANG
# ============================================================

def lihat_keranjang():
    header("KERANJANG BELANJA")
    if not KB["keranjang"]:
        print("  Keranjang kosong.")
        return

    total = 0
    for i, item in enumerate(KB["keranjang"], 1):
        subtotal = item["harga"] * item["jumlah"]
        total += subtotal
        print(f"  {i}. {item['name']}")
        print(f"     {format_harga(item['harga'])} x {item['jumlah']} = {format_harga(subtotal)}")
        separator("·")

    separator()
    print(f"  TOTAL: {format_harga(total)}")
    separator()


# ============================================================
# FITUR: CHECKOUT
# ============================================================

def checkout():
    header("CHECKOUT")
    if not KB["keranjang"]:
        print("  Keranjang kosong. Tambahkan produk terlebih dahulu.")
        return

    total = sum(item["harga"] * item["jumlah"] for item in KB["keranjang"])

    print("  Ringkasan pesanan:\n")
    for item in KB["keranjang"]:
        print(f"  • {item['name']} x{item['jumlah']} = {format_harga(item['harga'] * item['jumlah'])}")

    separator()
    print(f"  Total pembayaran: {format_harga(total)}")
    separator()

    konfirmasi = input("  Konfirmasi pembelian? (y/n): ").strip().lower()
    if konfirmasi != "y":
        print("  Checkout dibatalkan.")
        return

    # Kurangi stok
    for item in KB["keranjang"]:
        produk = get_entity(item["id"])
        if produk:
            produk["properties"]["stok"] -= item["jumlah"]

    # Simpan ke riwayat
    KB["riwayat_pembelian"].append({
        "items": KB["keranjang"].copy(),
        "total": total
    })

    KB["keranjang"].clear()
    print(f"\n  ✓ Pembelian berhasil! Total: {format_harga(total)}")
    print("  Terima kasih telah berbelanja di E-Commerce Platform.")
    separator()


# ============================================================
# FITUR: RIWAYAT PEMBELIAN
# ============================================================

def riwayat():
    header("RIWAYAT PEMBELIAN")
    if not KB["riwayat_pembelian"]:
        print("  Belum ada riwayat pembelian.")
        return

    for i, transaksi in enumerate(KB["riwayat_pembelian"], 1):
        print(f"  Transaksi #{i}")
        for item in transaksi["items"]:
            print(f"    • {item['name']} x{item['jumlah']} = {format_harga(item['harga'] * item['jumlah'])}")
        print(f"    Total: {format_harga(transaksi['total'])}")
        separator("·")


# ============================================================
# FITUR: RELASI SEMANTIK
# ============================================================

def lihat_relasi():
    header("RELASI SEMANTIK ENTITAS")
    print("  Pilih entitas:\n")

    entities = KB["entities"]
    for i, e in enumerate(entities, 1):
        print(f"  {i}. [{e['type'].upper()}] {e['name']}")

    separator()
    try:
        pilihan = int(input("  Pilih nomor (0 = batal): "))
        if pilihan == 0:
            return
        if not 1 <= pilihan <= len(entities):
            print("  Pilihan tidak valid.")
            return

        entity = entities[pilihan - 1]
        rels = get_relations_from(entity["id"])

        separator()
        print(f"  Entitas  : {entity['name']} [{entity['type'].upper()}]")
        print(f"  Deskripsi: {entity['description']}")
        separator("·")
        if rels:
            print("  Relasi semantik:\n")
            for r in rels:
                target = get_entity(r["to"])
                tname = target["name"] if target else r["to"]
                print(f"    → [{r['label']}] {tname}")
        else:
            print("  Tidak ada relasi keluar dari entitas ini.")
        separator()

    except ValueError:
        print("  Input tidak valid.")


# ============================================================
# MAIN MENU
# ============================================================

def main():
    menu = {
        "1": ("Lihat Semua Produk",       lihat_produk),
        "2": ("Cari Produk",              cari_produk),
        "3": ("Rekomendasi Semantik",     rekomendasi),
        "4": ("Tambah ke Keranjang",      tambah_keranjang),
        "5": ("Lihat Keranjang",          lihat_keranjang),
        "6": ("Checkout",                 checkout),
        "7": ("Riwayat Pembelian",        riwayat),
        "8": ("Lihat Relasi Semantik",    lihat_relasi),
        "0": ("Keluar",                   None),
    }

    print("\n" + "═" * 55)
    print("  🛒  E-COMMERCE PLATFORM — SEMANTIC KNOWLEDGE BASE")
    print("═" * 55)

    while True:
        print("\n  MENU UTAMA\n")
        for key, (label, _) in menu.items():
            print(f"  [{key}] {label}")

        separator()
        pilihan = input("  Pilih menu: ").strip()

        if pilihan == "0":
            print("\n  Sampai jumpa! 👋\n")
            break
        elif pilihan in menu:
            print()
            menu[pilihan][1]()
        else:
            print("  Menu tidak tersedia. Coba lagi.")


if __name__ == "__main__":
    main()
```
